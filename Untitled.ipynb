```python
# Importing all the essential libraries:
# - pandas & numpy for data handling and numerical operations
# - matplotlib for plotting time-series graphs
# - SARIMAX model from statsmodels for advanced forecasting
# - adfuller test to check if the data is stationary
# - ACF/PACF plots to understand autocorrelation patterns
# - mean_squared_error to evaluate forecast accuracy
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
from sklearn.metrics import mean_squared_error
```

```python
# Loading the dataset from a CSV file into a pandas DataFrame.
# This will store our monthly values so we can analyze and model the time series.
```

```python
df=pd.read_csv("Month_Value_1.csv")
```

```python
df.info()
```

```python
# Defining a helper function to run the Augmented Dickey-Fuller (ADF) test.
# It prints the p-value and tells us whether the given series is stationary.
# (p-value < 0.05 â†’ stationary, otherwise non-stationary)
```

```python
def adf_test(series,title):
    result=adfuller(series.dropna())
    print("ADF test for",title)
    print("p-value:",result[1])
    if result[1]<0.05:
        print(title,"is stationary")
    else:
        print(title,"is non stationary")
```

```python
df.head()
```

```python
df["Period"]=pd.to_datetime(df["Period"],format="%d.%m.%Y")
```

```python
df.set_index("Period",inplace=True)
```

```python
# Running the ADF test on the original Revenue series.
# This helps us check if the raw data is stationary before modeling.
```

```python
adf_test(df["Revenue"],"Original series")
```

```python
# Creating different versions of differenced data:
# - df_diff: first-order differencing to remove trend
# - df_seasonal_diff: seasonal differencing (lag 12) to remove yearly patterns
# - df_diff_seasonal: applying both first-order + seasonal differencing for full stationarity
```

```python
df_diff=df["Revenue"].diff(1)
df_seasonal_diff=df["Revenue"].diff(12).dropna()
df_diff_seasonal=df["Revenue"].diff(1).diff(12).dropna()
```

```python
# Running ADF tests on all differenced versions:
# - First difference: checks if removing trend helps
# - Seasonal difference: checks if removing yearly pattern helps
# - Combined: checks if using both makes the series fully stationary
```

```python
adf_test(df_diff,"First difference")
adf_test(df_seasonal_diff,"Seasonal difference")
adf_test(df_diff_seasonal,"Combined")
```

```python
# Applying log transformation to stabilize variance in the Revenue series.
# Then repeating the differencing steps (1st, seasonal, and combined) on the log-transformed data.
# Finally, checking stationarity again with ADF tests to confirm improvement after transformation.
```

```python
df["Revenue_log"]=np.log(df["Revenue"])
df_diff=df["Revenue_log"].diff(1)
df_seasonal_diff=df["Revenue_log"].diff(12).dropna()
df_diff_seasonal=df["Revenue_log"].diff(1).diff(12).dropna()
adf_test(df_diff,"First difference")
adf_test(df_seasonal_diff,"Seasonal difference")
adf_test(df_diff_seasonal,"Combined")
```

```python
# Plotting all versions of differenced data:
# - First difference
# - Seasonal difference
# - Combined difference
# This helps visually check whether the series looks stable and stationary after transformations.
```

```python

fig,axes=plt.subplots(3,1,figsize=(12,10))
df_diff.plot(ax=axes[0],title="First difference")
df_seasonal_diff.plot(ax=axes[1],title="Seasonal difference")
df_diff_seasonal.plot(ax=axes[2],title="Combined")
plt.tight_layout()
```

```python
# Plotting ACF and PACF of the fully differenced (trend + seasonality removed) series.
# These plots help identify the p, q, P, Q parameters for the SARIMAX model.
```

```python
fig,axes=plt.subplots(2,1,figsize=(12,10))
plot_acf(df_diff_seasonal,ax=axes[0],title="ACF for residuals")
plot_pacf(df_diff_seasonal,ax=axes[1],title="PACF for residuals")
```

```python
# Splitting the fully differenced series into train (80%) and test (20%) sets.
# This lets us train the model on past data and evaluate how well it predicts unseen data.
```

```python
train_size=int(len(df_diff_seasonal)*0.8)
train=df_diff_seasonal.iloc[:train_size]
test=df_diff_seasonal.iloc[train_size:]
```

```python
# Building a SARIMA model with selected parameters:
# - order=(1,1,1) handles trend patterns
# - seasonal_order=(1,1,1,12) captures yearly seasonality
# Fitting the model on training data and printing key coefficient details.
```

```python
sarima_model = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_fitted = sarima_model.fit(disp=False)
print("\n=== SARIMA Model Summary ===")
print(sarima_fitted.summary().tables[1])
```

```python
# Generating a 12-step (1 year) forecast using the trained SARIMA model.
# Extracting:
# - predicted_mean â†’ the actual forecast values
# - conf_int â†’ upper and lower confidence intervals for uncertainty estimation
```

```python
sarima_forecast = sarima_fitted.get_forecast(steps=12)
sarima_fc_mean = sarima_forecast.predicted_mean
sarima_fc_ci = sarima_forecast.conf_int()
```

```python
# Creating a simple seasonal naÃ¯ve forecast:
# It assumes values repeat exactly from 12 months ago.

# Comparing SARIMA vs naÃ¯ve forecast on the test set using RMSE.
# Lower RMSE = better model performance.
# Finally printing an insight showing whether SARIMA truly improves over the baseline.
```

```python
def seasonal_naive_forecast(series, steps):
    series = series.squeeze()  # ensure 1D
    naive_fc = series.shift(12).tail(steps)
    return naive_fc

naive_test_fc = seasonal_naive_forecast(train, len(test))
naive_next6 = seasonal_naive_forecast(df["Revenue_log"], 6)

# 7. Evaluation: RMSE on Test Set
sarima_test_rmse = np.sqrt(mean_squared_error(test, sarima_fc_mean[:len(test)]))
naive_test_rmse = np.sqrt(mean_squared_error(test, naive_test_fc))
print(f"\n=== Model Evaluation (Test Set RMSE) ===")
print(f"SARIMA RMSE: {sarima_test_rmse:.2f}")
print(f"Naive Seasonal RMSE: {naive_test_rmse:.2f}")
print("Insight: SARIMA outperforms naive by capturing trend/seasonality dynamics.")
```

```python
# Plotting the actual test data vs. SARIMA predictions and the seasonal naive baseline.
# Confidence intervals (CI) show uncertainty around SARIMA forecasts.
# This visual comparison helps evaluate how closely each model follows real values.
```

```python
plt.figure(figsize=(12, 6))
plt.plot(test.index, test.values, label='Actual Test', marker='o')
plt.plot(sarima_fc_mean[:len(test)].index, sarima_fc_mean[:len(test)].values, label='SARIMA Forecast', marker='x')
plt.plot(naive_test_fc.index, naive_test_fc.values, label='Naive Seasonal', marker='s')
plt.fill_between(sarima_fc_mean.index, sarima_fc_ci.iloc[:, 0], sarima_fc_ci.iloc[:, 1], color='gray', alpha=0.2, label='SARIMA 95% CI')
plt.title('Test Set: Actual vs Forecasts')
plt.xlabel('Date')
plt.ylabel('Sales ($ thousands)')
plt.legend()
plt.grid(True)
plt.show()
```
